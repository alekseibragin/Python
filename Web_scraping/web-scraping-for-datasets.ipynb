```python
from urllib.request import urlopen
```

```python
# Web scraping 
html = urlopen('https://ru.wikipedia.org/wiki/Deutsche_Bank').read().decode('utf-8')
print(html)
```

```python
# searching for a word 'Credit'
s = str(html)
ans = []
state = 0 
for c in s:
    if c == '<':
        state = 1
    if c == '>':
        state = 0
    elif state == 0:
        ans.append(c)
s = ''.join(ans)        
print(s.count('Credit'))
```

```python
# searching for a word 'Deposit'
print(s.count('Deposit'))
```

```python
# sorting table tags
from re import findall
redex = '<td>(.*?)</td>'
l = sorted(findall(redex,html))
l
```

```python
# counting the number of tags
from collections import Counter
d = Counter(l)
d
```

```python
list_d = list(d.items())
list_d
```

```python
# sorting by dictionary value
list_d.sort(key=lambda i: i[1])
list_d
```

```python
# search for links
html2 = urlopen('https://ru.wikipedia.org/wiki/Deutsche_Bank').read().decode('utf-8')
s = str(html2)
pos = s.find('<a href=')
while pos != -1:
    posquote = s.find('"', pos + 9)
    href = s[pos + 9:posquote]
    print(href)
    pos = html2.find('<a href=', pos + 1)
```

```python
# counting numbers on a page and operations on them
from bs4 import BeautifulSoup

resp = urlopen('https://ru.wikipedia.org/wiki/Deutsche_Bank') # скачиваем файл
html = resp.read().decode('utf8') # считываем содержимое
soup = BeautifulSoup(html, 'html.parser') # делаем суп
table = soup.find('table', attrs = {'class' : 'wikitable sortable'})
cnt = 0
for tr in soup.find_all('tr'):
    cnt += 1
    for td in tr.find_all(['td', 'th']):
        cnt *= 2
print(cnt)
```

```python
# search for all tags after a tag
#"https://stepik.org/media/attachments/lesson/209723/3.html"
html = urlopen('https://ru.wikipedia.org/wiki/Deutsche_Bank').read().decode('utf-8')
s = str(html)
soup = BeautifulSoup(s, 'html.parser') # делаем суп
for a in soup.find_all('a', href = True):
    print(a['href'])
```

```python
# counting all numbers in the table

html = urlopen('https://stepik.org/media/attachments/lesson/209723/5.html').read().decode('utf-8')
s = str(html)
soup = BeautifulSoup(s, 'html.parser') # делаем суп
sum = 0
for tag in soup.find_all('td'):
    sum += int(tag.string)
print(sum)
```
